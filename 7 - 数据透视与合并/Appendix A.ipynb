简而言之，append()的默认操作效果跟concat()相同， 都是实现两个DataFrame的纵向连接。事实上可以把它看做concat()的早期版本：

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 9, 2, 3],
)


df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)


df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "E": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 4, 10, 11],
)

df1
df2
df3
```

![](https://pandas.pydata.org/docs/_images/merging_concat_basic.png)

```python
frames = [df1, df2, df3]

result = pd.concat(frames)
```

```python
result # index 对此concat 无影响，column对此有影响,默认按列拼接
```

```python
result= pd.concat(frames, keys=["x", "y", "z"])
# pd.concat(frames, keys='abc')
```

# Database-style DataFrame or named Series joining/merging

## merge()
取两张表的键的交集，键是on的部分。

pd.merge(
    
    left,
    right,
    how="inner", 与concat 相反 outer
    on=None, 
    
on: Column or index level names to join on. `Must be found in both the left and right `DataFrame and/or Series objects. If not passed and left_index and right_index are False, the `intersection of the columns` in the DataFrames and/or Series will be inferred to be the join keys.
    
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)


```python
import  pandas as pd


left = pd.DataFrame(
    {
        "key": ["K1", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

left


right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K1", "K3"],
        "A": ["C0", "C1", "C2", "C3"],
        "B": ["D0", "D1", "D2", "D3"],
    }
)

right

result = pd.merge(left, right, on="key")
result
```

You can merge a mult-indexed Series and a DataFrame, if the names of the MultiIndex correspond to the columns from the DataFrame. Transform the Series to a DataFrame using Series.reset_index() before merging, as shown in the following example.


```python
df = pd.DataFrame({"Let": ["A", "B", "C"], "Num": [1, 2, 3]})
df
ser = pd.Series(
    ["a", "b", "c", "d", "e", "f"],
    index=pd.MultiIndex.from_arrays(
        [["A", "B", "C"] * 2, [1, 2, 3, 4, 5, 6]], names=["Let", "Num"]
    ),
)
ser
pd.merge(df, ser.reset_index(), on=["Let", "Num"])
```

```python
# Here is another example with duplicate join keys in DataFrames:


left = pd.DataFrame({"A": [1, 2], "B": [2, 2]})
left
right = pd.DataFrame({"A": [4, 5, 6], "B": [2, 2, 2]})
right

result = pd.merge(left, right, on="B", how="outer")
result
```
