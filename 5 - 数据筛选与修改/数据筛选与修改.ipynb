## 5-1 数据修改



### 1- <font color = '#FB8E00'>数据修改</font>｜列名

<br>

将原 `df` 列名 `Unnamed: 2`、`Unnamed: 3`、`Unnamed: 4` 修改为 `金牌数`、`银牌数`、`铜牌数`

### 4 - <font color = '#FB8E00'>数据修改</font>｜修改值

<br>

将 ROC（第一列第五行）修改为 俄奥委会

### 5 - <font color = '#FB8E00'>数据修改</font>｜替换值（单值）

<br>

将金牌数列的数字 `0` 替换为 `无`

```python
df['金牌数'].replace(0,'无',inplace=True)
```

### 6 - <font color = '#FB8E00'>数据修改</font>｜替换值（多值）



同时替换

- 将 `无` 替换为 缺失值
- 将 0 替换为 `None`

注意：缺失值的 Nan 该怎么生成？

```python
import numpy as np
df.replace(['无',0],[np.nan,'None'],inplace = True)
```

### 7 - 数据查看

查看各列数据类型

```python
df.dtypes
```

### 8 - <font color = '#FB8E00'>数据修改</font>｜修改类型

将 `金牌数` 列类型修改为 `int`

```python
df['金牌数'] = df['金牌数'].fillna('0').astype(int)
df['金牌数'] = df['金牌数'].fillna('0').map(lambda x :int(x))
```

### 9 - <font color = '#1B85FF' >数据增加</font>｜新增列（固定值）

<br>

### 10 - <font color = '#1B85FF' >数据增加</font>｜新增列（计算值）

<br>

**新增一列** 金银牌总数列，值为该国家金银牌总数

```python
df = df.replace('None',0)
df['金银牌总数']  = df['金牌数']  + df['银牌数']
df
```

### 11 - <font color = '#1B85FF' >数据增加</font>｜新增列（比较值）

新增一列 最多奖牌数量 列，值为该过金银牌数量种最多的一个奖牌数量

例如美国银牌最多，则为41，中国为38

```python
df['最多奖牌数量1'] = df.bfill(1)[["金牌数", "银牌数",'铜牌数']].mean(1) #{0 or ‘index’, 1 or ‘columns’} 

# bfill()与ffill()默认列前后填充 
df['最多奖牌数量2'] = df.ffill()[["金牌数", "银牌数",'铜牌数']].min(1) 
df['最多奖牌数量3'] = df.fillna(method='bfill')[["金牌数", "银牌数",'铜牌数']].min(1)
df['最多奖牌数量5']=df[["金牌数", "银牌数",'铜牌数']].max(axis=1)
df['最多奖牌数量6']=df[["金牌数", "银牌数",'铜牌数']].apply(max, axis=1)

(df[['A','B']].max(axis=1).max(axis=0)) #行列最大值
# 此方法中bfill与ffill一致
```

### 12 - <font color = '#1B85FF' >数据增加</font>｜新增列（判断值）

新增一列 金牌大于30

如果一个国家的金牌数大于 30 则值为 是，反之为 否

```python
df['金牌大于30']  = np.where(df['金牌数'] > 30, '是', '否')
df['金牌大于20']=df['金牌数'].apply(lambda x:  '是' if x >30  else '否' )
df
# lambda x: 语句1 if 条件1 else ( 语句2 if 条件2 else 语句3 )
bags_inv= {
    6: 'planes bag',
    4: 'pods bag',
    13: 'ground system bag'
}
df['金牌大于20']=df['金牌数'].map(bags_inv)
```

df.where从主体df出发，True返回df 本身的值，否则返回other的值；

np.where（cond，x，y）,True返回x的值，False返回y的值

```python
df['金牌大于30-02']  = np.where(df['金牌数'] > 30, '是', '否')
# df.where从主体df出发，True返回df 本身的值，否则返回other的值；np.where（cond，x，y）,True返回x的值，False返回y的值
```

### 13 - <font color = '#1B85FF' >数据增加</font>｜增加多列

新增两列，分别是

- 金铜牌总数（金牌数+铜牌数）
- 银铜牌总数（银牌数+铜牌数）

```python
df = df.assign(金铜牌总数=df.金牌数 + df.铜牌数,
         银铜牌总数=df.银牌数+df.铜牌数) 
```

### 14 - <font color = '#1B85FF' >数据增加</font>｜新增列（引用变量）

新增一列金牌占比，为各国金牌数除以总金牌数（gold_sum）

```python
df.eval(f'金牌占比 = 金牌数 / {gold_sum}',inplace=True)
```
